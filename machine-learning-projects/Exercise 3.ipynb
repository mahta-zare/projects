# ب

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(Standard, df1[279])
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn1 = KNeighborsClassifier(n_neighbors = 1, metric = 'euclidean')
knn1.fit(x_train,y_train)
y_prediction_test1 = knn1.predict(x_test)
y_prediction_train1 = knn1.predict(x_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import LabelBinarizer
```

```python
print(confusion_matrix(y_test, y_prediction_test1))
```

```python
print(classification_report(y_test, y_prediction_test1))
```

```python
L = LabelBinarizer()
L.fit(y_test)
y_test_roc1 = L.transform(y_test)
y_prediction_test_roc1 = L.transform(y_prediction_test1)
roc_auc_score(y_test_roc1, y_prediction_test_roc1)
```

```python
print(confusion_matrix(y_train, y_prediction_train1))
```

```python
print(classification_report(y_train, y_prediction_train1))
```

```python
L.fit(y_train)
y_train_roc1 = L.transform(y_train)
y_prediction_train_roc1 = L.transform(y_prediction_train1)
roc_auc_score(y_train_roc1, y_prediction_train_roc1)
```

```python
knn2 = KNeighborsClassifier(n_neighbors = 30, metric = 'euclidean')
knn2.fit(x_train,y_train)
y_prediction_test2 = knn2.predict(x_test)
y_prediction_train2 = knn2.predict(x_train)
```

```python
print(confusion_matrix(y_test, y_prediction_test2))
```

```python
print(classification_report(y_test, y_prediction_test2))
```

```python
L.fit(y_test)
y_test_roc2 = L.transform(y_test)
y_prediction_test_roc2 = L.transform(y_prediction_test2)
roc_auc_score(y_test_roc2, y_prediction_test_roc2)
```

```python
print(confusion_matrix(y_train, y_prediction_train2))
```

```python
print(classification_report(y_train, y_prediction_train2))
```

```python
L.fit(y_train)
y_train_roc2 = L.transform(y_train)
y_prediction_train_roc2 = L.transform(y_prediction_train2)
roc_auc_score(y_train_roc2, y_prediction_train_roc2)
```
