# Outlier

```python
df4
```

## 1. z-score

```python
labels = ['Population growth','Total population','Area (sq. km)','Coronavirus Cases']
```

```python
from scipy import stats
z=np.abs(stats.zscore(df4[labels]))
z
```

```python
print(np.where(z>3))
```

```python
z[0][2]
```

```python
z[14][1]
```

## 2. IQR score

```python
Q1=df4[labels].quantile(0.25)
Q1
```

```python
Q3=df4[labels].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
(df4[labels]<Q1-1.5*IQR) | (df4[labels]>Q3+1.5*IQR)
```

## Removing outliers

```python
df4.shape
```

```python
df7 = df4[(z<3).all(axis=1)]
df7.shape
```

```python
df8 = df4[((df4[labels]>=Q1-1.5*IQR) & (df4[labels]<=Q3+1.5*IQR)).all(axis=1)]
df8.shape
```
